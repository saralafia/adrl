## Demo Network Interactive Visualisation

```python
from pyvis import network as net
import chart_studio
import plotly.graph_objs as go
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

import networkx as nx
```

```python
df2_top_5 = pd.read_csv('df2_top_5.csv', sep=',')
```

```python
val_map_col = {'Chemistry': '#d1e4ed',
                  'Computer Science': '#5995bd',
                  'Education': '#d6f5d0',
                  'Electrical and Computer Engineering': '#809e7a',
                  'Geography': '#ed7a6f',
                  'History': '#ffdead',
                  'Materials': '#fcb377',
                  'Physics': '#8f7ca1',
                  'Psychology': '#fcffd6',
                  'Sociology': '#9c8667' 
                 }
```

```python
# Mapping the dictionary keys to the data frame. 
df2_top_5['colours'] = df2_top_5['Department'].map(val_map_col) 
df2_top_5['colours1'] = df2_top_5['Department1'].map(val_map_col) 

df2_top_5
```

```python
df2_top_5_nonNA = df2_top_5.dropna()
df2_top_5_nonNA
```

```python
df3_top_5 = df2_top_5_nonNA[0:775]
df3_top_5
```

```python
top_topic_net = net.Network(height="750px", width="100%", bgcolor="#222222", font_color="white", notebook=True)

# set the physics layout of the network
top_topic_net.barnes_hut()
top_topic_data = df3_top_5

sources = top_topic_data['title']
targets = top_topic_data['title1']
source_weight = top_topic_data['weight'].values*1000 
target_weight = top_topic_data['weight1'].values*1000 
source_department = top_topic_data['Department']
target_department = top_topic_data['Department1']
source_colour = top_topic_data['colours']
target_colour = top_topic_data['colours1']


edge_data = zip(sources, targets, source_weight, target_weight, source_department, target_department, source_colour, target_colour)

for e in edge_data:
    src = e[0]
    dst = e[1]
    src_w = e[2]
    tar_w = e[3]
    src_dep = e[4]
    tar_dep = e[5]
    src_col = e[6]
    tar_col = e[7]

    top_topic_net.add_node(src, src, title=src_dep, color = src_col, value = src_w+100000000)
    top_topic_net.add_node(dst, dst, title=tar_dep, color = tar_col, value = tar_w+100000000)
    top_topic_net.add_edge(src, dst, value=src_w*tar_w)
    
neighbor_map = top_topic_net.get_adj_list()

# add neighbor data to node hover data
for node in top_topic_net.nodes:
    node["title"] += " Neighbors:<br>" + "<br>".join(neighbor_map[node["id"]])
    node["value"] = len(neighbor_map[node["id"]])
    print(node["title"])

top_topic_net.show("demo_network.html")
```
